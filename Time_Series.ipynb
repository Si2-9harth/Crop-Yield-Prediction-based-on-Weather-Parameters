```python
def plot_series(time,series,format="-",start=0,end=None):
    plt.figure(figsize=(20,15))
    plt.plot(time[start:end],series[start:end],format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(False)
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Bidirectional
```

```python
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
```

```python
def predict_value(district,model):
  for i in range(48):
    x_input=district[-3:]
    x_input = x_input.reshape((1, n_steps, n_features))
    yhat = model.predict(x_input, verbose=1)
    print(yhat)
    district=np.append(district,yhat)

def model_predict():
  tf.keras.backend.clear_session()
  model = Sequential()
  model.add(LSTM(128, activation='relu', return_sequences=True, input_shape=(3, 1)))
  model.add(LSTM(128, activation='relu',return_sequences=True))
  model.add(LSTM(64, activation='relu'))
  model.add(Dense(64, activation='relu'))
  model.add(Dense(1))
  return model
```

```python
list_district=[balod, baloda_bazar, balrampur, bastar, bemetara,
       bijapur, bilaspur, dantewada,dhamtari, durg,
       gariyaband, janjgir_champa, jashpur, kabirdham, kanker,
       kondagaon, korba, korea, mahasamund, mungeli,
       narayanpur, raigarh, raipur, rajnandgaon, sukma,
       surajpur, surguja]
```

```python
for dis in list_district:
  raw_seq = dis
# choose a number of time steps
  n_steps = 3
# split into samples
  X, y = split_sequence(raw_seq, n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, features]
  n_features = 1
  tf.keras.backend.clear_session()
  X = X.reshape((X.shape[0], X.shape[1], n_features))
  model = model_predict()
  model.compile(optimizer='adam', loss='mse',metrics=['mae'])
  model.fit(X, y, epochs=200, verbose=0)
  predict_value(dis,model)
```
